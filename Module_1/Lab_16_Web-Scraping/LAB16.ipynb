#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
new_url  = 'https://github.com/trending/python?since=daily'
new_html = requests.get(new_url).content
sp = BeautifulSoup(new_html, 'html.parser')
print(sp.prettify())
````

````python
rep_selector = 'body > div.logged-in.env-production.page-responsive > div.application-main > main > div.position-relative.container-lg.p-responsive.pt-6 > div > div:nth-child(2) > article:nth-child(1) > h2 > a'
````

````python
rep_selector = 'h2 > a'
len(rep_selector)
````

````python
sp.select(rep_selector)[0].get_text()
````

````python
repo_names = []
for i in sp.select(rep_selector):
    repo_names.append(i.get_text().split())
len(repo_names)
````

````python
joined_list = []

for repo in repo_names:
    joined_list.append("".join(repo))

joined_list
````

````python
len(joined_list)
````
