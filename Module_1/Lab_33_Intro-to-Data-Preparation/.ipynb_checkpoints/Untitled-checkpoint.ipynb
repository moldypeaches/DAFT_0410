```python
sp = BeautifulSoup(response.content)
```

```python
#htlm= tag -- <div> class = 'abc' id='bla'
#</div>
#<a href >   = link
#<img href/>
```

```python
#container-59d6cdf5-b1fb-480b-9c6f-fd40bf23c7c0 > div:nth-child(2) > ul > li:nth-child(1) > div > div > a
```

```python
#div.container
tiles = sp.select('.fc-item__container')
```

```python
#select_one function
articles = []
for tile in tiles:
    title = tile.select_one('a').text.strip()
    link = tile.select_one('a')['href']
    article = {
        "title": title,
        "link": link,
    }
    articles.append(article)
```

```python
pd.DataFrame(articles)
```
