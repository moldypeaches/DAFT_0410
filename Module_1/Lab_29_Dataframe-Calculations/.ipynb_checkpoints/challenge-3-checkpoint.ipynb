```python
df2.reset_index()
```

```python
df2['agg_amount_spent'].quantile([0.75, 0.95])
```

```python
df2.loc[(df2["agg_amount_spent"] >= 1661.640) & (df2["agg_amount_spent"] <= 5840.182), 'customer_type'] = 'preferred_customer'
```

```python
df2.loc[df2["agg_amount_spent"] >= 5840.182, 'customer_type'] = 'VIP_customer'
```

## Q2: How to identify which country has the most VIP Customers?

```python
df_merged = df.merge(df2, how="inner", on="CustomerID")
```

```python
df_merged
```

```python
df_counts = df_merged.groupby("Country")["customer_type"].value_counts().to_frame("count").reset_index()

df_counts_VIP = df_counts[(df_counts['customer_type'] == 'VIP_customer') ]
```

```python
df_counts_VIP 
```

```python
df_counts_VIP.loc[df_counts_VIP['count'] == df_counts_VIP['count'].max(),['Country', 'count']]
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
df_counts.loc[(df_counts['customer_type'] == 'VIP_customer') | (df_counts['customer_type'] == 'preferred_customer')]
```
