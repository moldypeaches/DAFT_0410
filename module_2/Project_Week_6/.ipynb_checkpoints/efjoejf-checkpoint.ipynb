```python
pca = PCA(n_components=7)
```

```python
pca.fit(X_normalized)
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
X_pca = pca.transform(X_normalized)
```

```python
kmeans = KMeans(n_clusters=6, n_init=10)
```

```python
kmeans.fit(X_pca)
```

```python
cluster_ids = kmeans.predict(X_pca)
```

```python
cluster_distances = kmeans.transform(X_pca)
```

```python
for c in np.unique(cluster_ids):
    plt.scatter(X_pca[cluster_ids == c, 0], X_pca[cluster_ids == c, 1], marker='.', s=12)
```

```python
cluster_ids = KMeans(n_clusters=4, n_init=10).fit_predict(X_normalized)
```

```python
for c in np.unique(cluster_ids):
    plt.scatter(X_normalized[cluster_ids == c, 0], X_normalized[cluster_ids == c, 1], marker='.', s=12)
```

```python
cluster_ids = KMeans(n_clusters=3, n_init=10).fit_predict(X_normalized)
```

```python
for c in np.unique(cluster_ids):
    plt.scatter(X_normalized[cluster_ids == c, 0], X_normalized[cluster_ids == c, 1], marker='.', s=12)
```
