```python
import pandas as pd
import numpy as np
from plotly import express as px
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('titanic.csv', sep=';')
df.dropna(subset=["Age", "Fare", "Pclass", "Sex"], inplace=True)
```

```python
df["Male"] = (df["Sex"] == "male") * 1
df["Female"] = (df["Sex"] == "female") * 1
```

```python
scaler = StandardScaler()
X_normalized = scaler.fit_transform(df[["Age", "Fare", "Pclass", "Male", "Female"]])
```

```python
df_normalized = pd.DataFrame(X_normalized, columns=["Age", "Fare", "Pclass", "Male", "Female", "Female"])
```

```python
pca = PCA(n_components=5)
```

```python
pca.fit(X_normalized)
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
X_pca = pca.transform(X_normalized)
```

```python
kmeans = KMeans(n_clusters=8, n_init=10)
```

```python
kmeans.fit(X_pca)
```

```python
cluster_ids = kmeans.predict(X_pca)
```

```python
cluster_distances = kmeans.transform(X_pca)
```

```python
for c in np.unique(cluster_ids):
    plt.scatter(X_pca[cluster_ids == c, 0], X_pca[cluster_ids == c, 1], marker='.', s=12)
```

```python
cluster_ids = KMeans(n_clusters=6, n_init=10).fit_predict(X_normalized)
```

```python
for c in np.unique(cluster_ids):
    plt.scatter(X_normalized[cluster_ids == c, 0], X_normalized[cluster_ids == c, 1], marker='.', s=12)
```
